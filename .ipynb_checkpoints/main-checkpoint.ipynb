```python
#convert class dataset => yolo_dataset
import shutil, random
from pathlib import Path
from PIL import Image
import os, yaml

PROJECT_ROOT = Path.cwd()
DATA_DIR = PROJECT_ROOT / "data"
TRAIN_DIR = DATA_DIR / "Train"
TEST_DIR  = DATA_DIR / "Test"
OUT = PROJECT_ROOT / "yolo_dataset"
IMG_OUT_TRAIN = OUT / "images" / "train"
IMG_OUT_VAL   = OUT / "images" / "val"
LBL_OUT_TRAIN = OUT / "labels" / "train"
LBL_OUT_VAL   = OUT / "labels" / "val"
EXTS = {".jpg", ".jpeg", ".png", ".ppm", ".bmp"}

# dossier creation
for p in [IMG_OUT_TRAIN, IMG_OUT_VAL, LBL_OUT_TRAIN, LBL_OUT_VAL]:
    p.mkdir(parents=True, exist_ok=True)

# class list depuis train
class_dirs = sorted([p for p in TRAIN_DIR.iterdir() if p.is_dir()], key=lambda x: int(x.name) if x.name.isdigit() else x.name)
class_names = [p.name for p in class_dirs]
print("Detected classes (len={}): {}".format(len(class_names), class_names[:10]))

# copier images
def write_label_and_copy(src_image_path, dest_img_path, dest_lbl_path, class_id):
    img = Image.open(src_image_path)
    w,h = img.size
    # full image bbox: center (0.5,0.5), width=1.0, height=1.0
    line = f"{class_id} {0.5:.6f} {0.5:.6f} {1.0:.6f} {1.0:.6f}\n"
    shutil.copy(src_image_path, dest_img_path)
    with open(dest_lbl_path, "w") as f:
        f.write(line)

#proc train
for cls in class_names:
    src_cls_dir = TRAIN_DIR / cls
    for img_path in src_cls_dir.iterdir():
        if img_path.suffix.lower() not in EXTS: 
            continue
        dest_img = IMG_OUT_TRAIN / img_path.name
        dest_lbl = LBL_OUT_TRAIN / (img_path.stem + ".txt")
        write_label_and_copy(img_path, dest_img, dest_lbl, class_names.index(cls))

#proc test
# If Test a sous dossier copier depuis chacune, else copier images
if TEST_DIR.exists():
    #verfier sous dossiers
    test_subdirs = [p for p in TEST_DIR.iterdir() if p.is_dir()]
    if test_subdirs:
        for sub in test_subdirs:
            for img_path in sub.iterdir():
                if img_path.suffix.lower() not in EXTS: continue
                dest_img = IMG_OUT_VAL / img_path.name
                dest_lbl = LBL_OUT_VAL / (img_path.stem + ".txt")
                # if classes in Test meme nom dans Train utilise meme classe index, else map
                cls_name = sub.name
                if cls_name in class_names:
                    cls_id = class_names.index(cls_name)
                else:
                    #si pas trouver skip ou map 0
                    cls_id = 0
                write_label_and_copy(img_path, dest_img, dest_lbl, cls_id)
    else:
        # images directement dans test 
        for img_path in TEST_DIR.iterdir():
            if img_path.suffix.lower() not in EXTS: continue
            dest_img = IMG_OUT_VAL / img_path.name
            dest_lbl = LBL_OUT_VAL / (img_path.stem + ".txt")
            write_label_and_copy(img_path, dest_img, dest_lbl, 0)

# ecrire data.yaml
data_yaml = {
    "path": str(OUT),
    "train": "images/train",
    "val": "images/val",
    "names": class_names
}
with open(OUT / "data.yaml", "w") as f:
    yaml.dump(data_yaml, f, sort_keys=False)
print("yolo_dataset created at:", OUT)
print("data.yaml classes:", len(class_names))
```

```python
# affiche quelque full image bbox
import random
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

OUT = PROJECT_ROOT / "yolo_dataset"
img_dir = OUT / "images" / "train"
lbl_dir = OUT / "labels" / "train"

imgs = [p for p in img_dir.iterdir() if p.suffix.lower() in {".jpg",".jpeg",".png",".ppm",".bmp"}]
sample = random.sample(imgs, min(6, len(imgs)))
cols = 3
rows = (len(sample)+cols-1)//cols
fig, axs = plt.subplots(rows, cols, figsize=(5*cols,4*rows))
axs = axs.flatten()
for ax in axs[len(sample):]:
    ax.axis('off')
for ax,p in zip(axs, sample):
    im = Image.open(p)
    ax.imshow(im)
    lbl = lbl_dir/(p.stem + ".txt")
    if lbl.exists(): 
        for line in lbl.read_text().splitlines():
            if not line.strip(): continue
            cid, xc, yc, w, h = map(float, line.split())
            iw, ih = im.size
            x1 = (xc - w/2) * iw
            y1 = (yc - h/2) * ih
            rect = Rectangle((x1,y1), w*iw, h*ih, edgecolor='r', facecolor='none', linewidth=2)
            ax.add_patch(rect)
            ax.text(x1, y1-5, str(int(cid)), color='yellow')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
from ultralytics import YOLO
from pathlib import Path
import torch, json

# --------------------- CONFIG ---------------------
RUN_NAME = "gtsrb_quick"
RUN_DIR = Path("runs/detect") / RUN_NAME
LAST = RUN_DIR / "weights" / "last.pt"        # dernier checkpoint YOLO
STATE = RUN_DIR / "train_state.json"          # notre compteur depochs
DATA = "yolo_dataset/data.yaml"               # dataset YOLO

TARGET_EPOCHS = 15  # nombre total depochs souhaites
# --------------------------------------------------

# ---- 1) Lire combien depochs ont dj t faites ----
prev = 0

# depuis notre fichier dtat
if STATE.exists():
    prev = json.loads(STATE.read_text()).get("epochs_completed", 0)

# fallback : lire epoch depuis last.pt (si prsent et si YOLO la sauvegard)
if LAST.exists():
    ckpt = torch.load(LAST, map_location="cpu")
    if "epoch" in ckpt:
        prev = max(prev, ckpt["epoch"] + 1)

print("poques dj compltes :", prev)

# ---- 2) Charger le modle depuis last.pt ou repartir de zro ----
model = YOLO(str(LAST)) if LAST.exists() else YOLO("yolov8n.pt")

# ---- 3) Entraner 1 epoch  la fois jusqu TARGET_EPOCHS ----
while prev < TARGET_EPOCHS:
    print(f"\nExcution de l'poque {prev+1}/{TARGET_EPOCHS}")

    model.train(
        data=DATA,
        epochs=1,          # exactement 1 epoch
        imgsz=320,
        batch=4,
        name=RUN_NAME,
        device=0,          # GPU = 0
        exist_ok=True,
        resume=False,      # important : pas de resume YOLO
        cache=True,
        workers=0,
        mosaic=0.0,
        augment=False
    )

    # recharger last.pt avant lepoch suivante
    if LAST.exists():
        model = YOLO(str(LAST))

    # mettre  jour le compteur
    prev += 1
    STATE.write_text(json.dumps({"epochs_completed": prev}))
    print("Compteur mis  jour :", prev)

print("\n Entranement termin !")
```

```python
#inference webcam (q pour quitter)
import cv2, time
from ultralytics import YOLO

weights = "runs/detect/gtsrb_yolov8n_quick/weights/best.pt"
if not Path(weights).exists():
    weights = "yolov8n.pt"
yolo = YOLO(weights)

cap = cv2.VideoCapture(0)   # 0 pour webcam par defaut /// ou video path
if not cap.isOpened():
    raise RuntimeError("Cannot open webcam. Use video file path if headless.")

fps_smooth = 0.0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    t0 = time.time()
    res = yolo(frame, imgsz=320)[0]
    for box in res.boxes:
        x1,y1,x2,y2 = map(int, box.xyxy[0].tolist())
        conf = float(box.conf[0]); cls = int(box.cls[0])
        label = f"{yolo.model.names[cls]} {conf:.2f}"
        cv2.rectangle(frame, (x1,y1),(x2,y2),(0,255,0),2)
        cv2.putText(frame, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0),2)
    fps = 1.0/(time.time()-t0)
    fps_smooth = 0.9*fps_smooth + 0.1*fps if fps_smooth else fps
    cv2.putText(frame, f"FPS: {fps_smooth:.1f}", (10,30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,255,255),2)
    cv2.imshow("YOLOv8 GTSRB", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
